```python
import nflgame
import time
import csv
import pandas as pd
import time
import re
import numpy as np

```

```python
with open('stats.csv', 'rb') as f:
    stats = f.read()
    

stat_list = stats.split(',')

    
```

```python
stat_list
```

```python
stat_frame = pd.DataFrame(columns=stat_list)
```

```python
for year in range(2009,2016):
    
    print "Collecting data from " + str(year) + "..." 
    filename = 'players_season' + str(year) + '.csv'
    f = open(filename, 'w')
    weeks = 17
    lines = []  # init holder of all player stat lines to sort
    running_length = 0  # line counter for adding the correct week to the proper lines
    stat_frame = pd.DataFrame(columns=stat_list)

    # there are 17 weeks in an NFL season, each player can have max 16 stat lines due to bye week
    for week in range(1,weeks + 1): 
        # combine_max_stats sums stats over weeks, so we do one week at a time
        # we choose to do this because combine_max_stats gives us more stat categories than other methods
        nflgame.combine_max_stats(nflgame.games(year,[week])).csv('temp.csv', allfields = True) 

        # no way to include append in nflgame .csv method, so we do it in a roundabout fashion
        max_stats =pd.read_csv('temp.csv', index_col=None)
        
        red_stats = max_stats.loc[:,stat_list]
        
        red_stats['year'] = np.repeat(year, red_stats.shape[0])
        red_stats['week'] = np.repeat(week, red_stats.shape[0])
        
        stat_frame = pd.concat([stat_frame, red_stats])
        
        
        del red_stats
        del max_stats
        
        
        
        
        
        
```

```python
stat_list.insert(0, 'week')
stat_list.insert(0, 'year')
```

```python
stat_frame2 = stat_frame.loc[:, stat_list]
```

```python
stat_frame2.sort_values(by=['year', 'name', 'week'], inplace = True)
stat_frame2
```
